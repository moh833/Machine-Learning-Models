```python
import sklearn
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
from sklearn import linear_model, preprocessing
```

```python
data = pd.read_csv('car.data')
data = sklearn.utils.shuffle(data)
data.head()
```

```python
le = preprocessing.LabelEncoder()
```

```python
buying = le.fit_transform(data['buying'])
maint = le.fit_transform(data['maint'])
doors = le.fit_transform(data['doors'])
persons = le.fit_transform(data['persons'])
lug_boot = le.fit_transform(data['lug_boot'])
safety = le.fit_transform(data['safety'])
cls = le.fit_transform(data['class'])
```

```python
predict = 'class'
```

```python
X = np.column_stack((buying, maint, doors, persons, lug_boot, safety))
y = cls
```

```python
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.1)
```

```python
model = KNeighborsClassifier(n_neighbors=8)
```

```python
model.fit(x_train, y_train)
```

```python
acc = model.score(x_test, y_test)
acc
```

```python
predictions = model.predict(x_test)
```

```python
names = ['unacc', 'acc', 'good', 'vgood'] 
for x in range(len(predictions)):
    print('Predicted:', names[predictions[x]], 'Data:', x_test[x], 'Actual:', names[y_test[x]])
```

```python
for x in range(len(predictions)):
    print(f'8-neighbors of point {x_test[x]} -> {model.kneighbors([x_test[x]], 8)}')
    print()
```
